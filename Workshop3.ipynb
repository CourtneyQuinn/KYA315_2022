# Velocity potential

In this notebook we explore an example of plotting velocity potential for fluid flows

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Consider 4a from assignment 1

$$ \mathbf{u} = x\hat{\mathbf{i}}-y\hat{\mathbf{j}} $$

In vector form this is $\mathbf{u}=\begin{bmatrix} x \\ -y \end{bmatrix}$

```python
def velocity(x,y):
    u = x
    v = -y
    U = np.array([u,v])
    return U
```

```python
velocity(0,1)
```

## Calculate velocity on gridded domain

```python
x = np.linspace(-3,3,25)
y = np.linspace(-3,3,25)
X,Y = np.meshgrid(x,y)
```

```python
U = velocity(X,Y)
```

```python
U.shape
```

```python
u = U[0,:,:]
v = U[1,:,:]
```

```python
u.shape
```

```python
fig = plt.figure(figsize=(5,5))
plt.quiver(X,Y,u,v);
```

## Define velocity potential

```python
def potential(x,y):
    Phi = 1/2*(x**2-y**2)
    return Phi
```

```python
Phi = potential(X,Y)
```

```python
Phi.shape
```

## Plot velocity potential

```python
fig = plt.figure(figsize=(5,5))
p = plt.contourf(X,Y,Phi,cmap='RdBu',levels=100)
plt.colorbar(p)
```

## Plot streamlines

```python
fig = plt.figure(figsize=(5,5))
plt.streamplot(X,Y,u,v, density=0.5);
```

```python
fig = plt.figure(figsize=(5,5))
p = plt.contourf(X,Y,Phi,cmap='RdBu',levels=100)
plt.streamplot(X,Y,u,v, density=0.5,color='k');
plt.colorbar(p)
```
