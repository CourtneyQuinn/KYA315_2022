# Streamlines and velocity fields

In this notebook we explore examples of calculating and plotting streamlines and streak lines for fluid flows

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Consider the example from last week's workshop

$$ \mathbf{u} = \frac{x\hat{\mathbf{j}}-y\hat{\mathbf{i}}}{x^2+y^2} $$

In vector form this is $\mathbf{u}=\begin{bmatrix} \frac{-y}{x^2+y^2} \\ \frac{x}{x^2+y^2} \end{bmatrix}$

```python
def velocity(x,y):
    u = -y/(x**2+y**2)
    v = x/(x**2+y**2)
    U = np.array([u,v])
    return U
```

```python
velocity(0,1)
```

## Gridded domain

To plot velocity field we need to create a griddied domain

```python
x = np.linspace(-3,3,25)
y = np.linspace(-3,3,25)
X,Y = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

```python
plt.plot(X[:,0],Y[:,0],'.')
```

```python
plt.plot(X[0,:],Y[0,:],'.')
```

## Calculate velocity on gridded domain

```python
U = velocity(X,Y)
```

```python
U.shape
```

```python
u = U[0,:,:]
v = U[1,:,:]
```

```python
u.shape
```

```python
fig = plt.figure(figsize=(8,8))
plt.quiver(X,Y,u,v);
```

## Plot streamlines

```python
fig = plt.figure(figsize=(8,8))
plt.streamplot(X,Y,u,v, density=0.5);
```
